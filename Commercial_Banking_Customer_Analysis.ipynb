Handling missing value

#2)Handling missing value

```python
df.fillna(method='ffill')
```

```python
df['credit_score'] = df['credit_score'].fillna(
    df['credit_score'].median()
)
```

#3) Remove duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop_duplicates()
```

#4) Standardize Column Names

```python
df.columns = df.columns.str.lower()
```

```python
bins=[18,25,35,50,65]

labels=['Young','Adult',
        'Mid-Career','Senior']

df['age_group']=pd.cut(
    df['age'],
    bins=bins,
    labels=labels
)
```

#5) Revenue by city

```python
df.groupby('city')['transaction_amount'].sum()
```
